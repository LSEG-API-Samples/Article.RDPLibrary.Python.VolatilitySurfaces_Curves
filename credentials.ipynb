# Refinitiv API Library Examples - Helpers

This script provides helpers for other Jupyter Notebook examples that demonstrate the Refinitiv API Library

```python
import getpass
```

```python
def get_app_key():
    try:
        app_key = %env APP_KEY
    except:
        app_key = input('Please input App Key:')        
    return app_key
```

```python
def get_rdp_login():
    try:
        rdp_login = %env RDP_LOGIN
    except:
        rdp_login = input('Please input Data Platform login:')        
    return rdp_login
```

```python
def get_rdp_password():
    try:
        rdp_password = %env RDP_PASSWORD
    except:
        rdp_password = getpass.getpass('Please input Data Platform password:')        
    return rdp_password
```
